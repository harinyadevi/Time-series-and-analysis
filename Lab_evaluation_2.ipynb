```python
wt = pd.read_csv("/content/Weight.csv")
```

```python
wt
```

```python
#moving average for smoothing

from matplotlib import pyplot

rolling = wt.rolling(window=3)
rolling_mean = rolling.mean()
print(rolling_mean.head(10))
```

```python
# plot original and transformed dataset
wt.plot()
rolling_mean.plot(color='red')
pyplot.show()
```

```python
import seaborn as sns
sns.lineplot( x = 'Date', y = 'Weights', data = wt, label = 'OriginalDataset')
  
# plot of transformed data
sns.lineplot( x = 'Date', y = 'Weights', data = rolling_mean, label = 'RollingMeanTransformedData')
```
